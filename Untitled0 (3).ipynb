```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
training_data = pd.read_csv('train.csv')
training_data.head()
```

```python
housing_data = training_data[['GrLivArea','BedroomAbvGr','SalePrice']]
housing_data.dropna(inplace=True)
```

```python
features = housing_data[['GrLivArea', 'BedroomAbvGr']]
target = housing_data['SalePrice']

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.2, random_state=42)
```

```python
linear_regression_estimator = LinearRegression()
# Use the correctly named training feature and target variables
linear_regression_estimator.fit(features_train, target_train)
```

```python
# Use the correctly named training feature and target variables
linear_regression_estimator.fit(features_train, target_train)

# Use the correct variable name for the trained model
predicted_prices = linear_regression_estimator.predict(features_test)
price_mse = mean_squared_error(target_test, predicted_prices)
price_r2 = r2_score(target_test, predicted_prices)
print("Housing Price MSE:", price_mse)
print("Housing Price RÂ²:", price_r2)
```

```python
test_data = pd.read_csv('test.csv')
final_features = test_data[['GrLivArea','BedroomAbvGr']]
# Use the correct variable name for the trained model
price_predictions = linear_regression_estimator.predict(final_features)
result_file = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': price_predictions})
result_file.to_csv('house_price_predictions.csv', index=False)
print("Prediction file successfully generated.")
```

```python
plt.scatter(target_test, predicted_prices)
plt.xlabel('Actual Home Values ($)')
plt.ylabel('Model Predictions ($)')
plt.title("True Values vs Model Predictions")
plt.grid(True, alpha=0.3)
plt.show()
```
